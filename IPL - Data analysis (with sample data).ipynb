# IPL - Data Analysis
# MATRICES / NUMPY

- A **matrix** is a tabular (rows and columns) representation of data.  
- Many datasets are stored in table format, making matrices a **very important concept in Python**, especially in data analysis.  
- Since we frequently work with **dataframes**, understanding matrices is crucial.  
- A **list** is one-dimensional, whereas a **matrix** is multi-dimensional.  
- **Indexing** is essential when plotting or analyzing specific data points.

- Identify **trends** and **patterns** in their performance.

# xticks()
-xticks refers to the locations and labels of ticks on the x-axis of a plot.

```python
plt.plot(Salary[0], c = 'Green', ls = '--', marker = 's', ms = '5')
plt.xticks(list(range(0,10)), Seasons)
plt.show()
```

```python
# you are assigning the name of the first player (from a list called Players) to a variable called label.
plt.plot(Salary[0], c='Green', ls = ':', marker = 's', ms = 7, label = Players[0])
plt.xticks(list(range(0,10)), Seasons,rotation='vertical') # rotation - rotating the years to vertical
plt.show()
```

# legend()
- A legend tells you what each line or marker in the plot represents.

```python
plt.plot(Salary[0], c='Green', ls = '--', marker = 's', ms = 7, label = Players[0])
plt.plot(Salary[1], c='Blue', ls = ':', marker = 'o', ms = 5, label = Players[1])
plt.plot(Salary[2], c='purple', ls = '-', marker = '^', ms = 8, label = Players[2])
plt.plot(Salary[3], c='Red', ls = '--', marker = 'd', ms = 8, label = Players[3])
plt.legend() 
plt.xticks(list(range(0,10)), Seasons,rotation='vertical')

plt.show()
```
