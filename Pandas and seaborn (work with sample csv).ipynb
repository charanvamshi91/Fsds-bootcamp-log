# Pandas Library 
- Pandas is a powerful and easy-to-use Python library for data manipulation and analysis. It's built on top of NumPy and is especially useful for working with structured data like tables (CSV, Excel, SQL, etc.).

# Difference b/w Pandas and Numpy
> NumPy is mainly used for numerical calculations with arrays and matrices, and it works best with homogeneous (same type) data.
It is fast and ideal for scientific computing and mathematical operations.

> Pandas is used for data analysis and manipulation, especially with structured (tabular) data like Excel or CSV files.
>It can handle different data types in each column and is easier to use for real-world data tasks like filtering, cleaning, and grouping.

-In short:
- NumPy = best for math with arrays
- Pandas = best for working with data tables

# Seaborn Library
- Seaborn is a Python library used for creating attractive and easy statistical visualizations, built on top of Matplotlib and works well with Pandas DataFrames.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 5
import warnings
warnings.filterwarnings('ignore')
```

```python
stats
```

```python
stats['InternetUsers']
```

```python
# visualization through displot
vis_1 = sns.displot(stats['InternetUsers']) # THIS ANALYSIS IS CALLED UNIVERIATE ANALYSIS
plt.show()
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins = 10) 
plt.show()
```

```python
# visualization through Box Plot
vis_2 = sns.boxplot(data = stats, x = "IncomeGroup", y = "BirthRate", hue = 'IncomeGroup' ) # THIS ANALYSIS IS CALLED BI - VARIATE ANALYSIS
plt.show()
```

```python
vis_3 = sns.lmplot(data = stats, x = 'InternetUsers', y = 'BirthRate')
plt.show()
```

```python
vis_4 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate', fit_reg = False ,hue = 'IncomeGroup')
plt.show()
```
